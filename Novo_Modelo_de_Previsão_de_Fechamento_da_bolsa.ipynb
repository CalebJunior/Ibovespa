```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.graph_objects as go
```

```python
df = pd.read_csv('/content/Dados_Historicos_Ibovespa.csv',parse_dates=['Data'])
df.index = pd.to_datetime(df.index)
```

```python

# Definindo a coluna 'Data' como índice
df.set_index('Data', inplace=True)
```

```python
import plotly.graph_objects as go

# Criar o gráfico
fig = go.Figure()

# Adicionar a linha do DataFrame
fig.add_trace(go.Scatter(x=df.index, y=df['Último'],
                         mode='lines',
                         name='Preço',
                         line=dict(width=2)))

# Atualizar o layout do gráfico
fig.update_layout(
    title='Evolução do Fechamento da Bolsa nos últimos 5 anos',
    xaxis_title='Data',
    yaxis_title='Preço',
    xaxis=dict(
        tickformat='%Y-%m-%d',  # Formato de data
        tickangle=-45,  # Rotacionar os rótulos do eixo x
    ),
    yaxis=dict(
        title='Preço',
    ),
    legend_title='Legenda',
    autosize=True,
    template='plotly_white'  # Escolher um template para a aparência do gráfico
)

# Mostrar o gráfico
fig.show()
```

```python
# Supondo que 'test' seja o DataFrame de teste e 'Previsão' seja a coluna com as previsões

train = df.loc[df.index < '2024-01-01']
test = df.loc[df.index >= '2024-01-01']

# Gerando previsões fictícias para fins de exemplo
np.random.seed(0)
test['Previsão'] = test['Último'] + np.random.normal(0, 1, size=len(test))

# Calculando o WMAPE (Weighted Mean Absolute Percentage Error)
wmape = np.sum(np.abs(test['Último'] - test['Previsão'])) / np.sum(test['Último']) * 100

# Imprimindo a taxa de erro
print(f'WMAPE: {wmape:.2f}%')

# Criando o gráfico com Plotly
fig = go.Figure()

# Adicionando a série de dados reais
fig.add_trace(go.Scatter(
    x=test.index,
    y=test['Último'],
    mode='lines',
    name='Real'
))

# Adicionando a série de previsões
fig.add_trace(go.Scatter(
    x=test.index,
    y=test['Previsão'],
    mode='lines',
    name='Previsão'
))

# Configurando o layout
fig.update_layout(
    title='Resultados Reais vs. Previsões do Fechamento da bolsa',
    xaxis_title='Date',
    legend_title='Legenda',
    width=800,
    height=600,
    xaxis_rangeslider_visible=False,  # Remove o controle deslizante de intervalo se não for necessário
    xaxis=dict(
        rangeselector=dict(
            buttons=list([
                dict(count=1,
                     label='1M',
                     step='month',
                     stepmode='backward'),
                dict(count=6,
                     label='6M',
                     step='month',
                     stepmode='backward'),
                dict(count=1,
                     label='Ano',
                     step='year',
                     stepmode='todate'),
                dict(label='Todos',
                     step='all')
            ])
        ),
        rangeslider=dict(
            visible=True
        ),
        type='date'
    ),
    yaxis=dict(
        tickformat='.3f'  # Formata os ticks do eixo y para 3 casas decimais
    )
)

# Exibindo o gráfico
fig.show()
```

```python


# Supondo que o DataFrame df já tenha a coluna 'Data' como índice
# Dividindo os dados em treino e teste
train = df.loc[df.index < '2024-01-01']
test = df.loc[df.index >= '2024-01-01']

# Gerando previsões fictícias para fins de exemplo
# Substitua isto pela previsão real do seu modelo
np.random.seed(0)
test['Previsão'] = test['Último'] + np.random.normal(0, 1, size=len(test))

# Calculando o WMAPE (Weighted Mean Absolute Percentage Error)
wmape = np.sum(np.abs(test['Último'] - test['Previsão'])) / np.sum(test['Último']) * 100

# Arredondando os valores das colunas 'Último' e 'Previsão' para 3 casas decimais
test['Último'] = test['Último'].round(3)
test['Previsão'] = test['Previsão'].round(3)

# Calculando o erro percentual absoluto para cada previsão e formatando como percentual
test['Erro Percentual'] = (np.abs(test['Último'] - test['Previsão']) / test['Último'] * 100).round(2)

# Garantindo que as colunas 'Último' e 'Previsão' tenham exatamente 3 casas decimais
test['Último'] = test['Último'].apply(lambda x: f'{x:.3f}')
test['Previsão'] = test['Previsão'].apply(lambda x: f'{x:.3f}')

# Criando a tabela para visualizar os resultados reais, previsões e erro
result_table = test[['Último', 'Previsão', 'Erro Percentual']].copy()
result_table.reset_index(inplace=True)
result_table.rename(columns={'index': 'Date', 'Último': 'Fechamento do Dia'}, inplace=True)


# Formatação da tabela para exibir valores em percentual
result_table['Erro Percentual'] = result_table['Erro Percentual'].apply(lambda x: f'{x:.2f}%')

# Criando a tabela interativa com Plotly
result_table_fig = go.Figure(data=[go.Table(
    header=dict(values=list(result_table.columns),
                fill_color='paleturquoise',
                align='left'),
    cells=dict(values=[result_table[col] for col in result_table.columns],
               fill_color='lavender',
               align='left'))
])

# Configurando o layout para centralizar o título
result_table_fig.update_layout(
    title={
        'text': 'Tabela Resultados Reais vs. Previsões com Taxa de Erro',
        'x': 0.5,  # Centralizar horizontalmente
        'xanchor': 'center',  # Âncora do título no centro
        'font': {'size': 20}  # Tamanho da fonte do título
    },
    width=1000,
    height=600
)

# Imprimindo a taxa de erro
print(f'WMAPE: {wmape:.2f}%')

# Exibindo a tabela
result_table_fig.show()
```
